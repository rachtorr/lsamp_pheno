```python
import ee
import geemap
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#!pip install contextily
#from contextily import Place

# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize(project='sound-octagon-449519-e8')
```

```python
neon = ee.Geometry.MultiPoint(
        [[-119.257549, 37.059719], #BIGC
         [-119.26219, 37.03337]]); #SOAP

snow = ee.Geometry.MultiPoint(
      [[-119.22, 37.23], # Huntinhton Lake
       [-119.2, 37.16]]) # Tamarack Summit
```

```python
wshed = ee.FeatureCollection('USGS/WBD/2017/HUC08')
dem = ee.Image('USGS/3DEP/10m')

# Create a map object
m = geemap.Map()

wshedPoints1 = wshed.filterBounds(neon)
wshedPoints2 = wshed.filterBounds(snow)

elevation = dem.select('elevation').clipToCollection(wshedPoints1)
elevation2 = dem.select('elevation').clipToCollection(wshedPoints2)

terrain_pal = {'min': 0, 'max': 3000,   'palette': [
    '#00A600',  # dark green (low elevation)
    '#63C600',  # green
    '#E6E600',  # yellow-green
    '#E9BD3A',  # light yellow-brown
    '#ECB176',  # light brown
    '#EFC2B3',  # pink-brown
    '#F2F2F2',  # white
    '#B4D0E7',  # light blue
    '#729ECE',  # medium blue
    '#2F8BC9',  # blue
    '#000080'   # dark blue (high elevation)
  ],
}

m.addLayer(elevation, terrain_pal, 'elevation_neon');
m.addLayer(elevation2, terrain_pal, 'elevation_snow');

style = {'color': 'e82c35ff', 'pointSize': 3, 'pointShape': 'circle', 'width': 2, 'lineType': 'solid', 'fillColor': '000000a8'}
m.addLayer(neon, style, 'NEON')
m.addLayer(snow, None, 'SWE')

# Add a layer control panel
m.add_layer_control()

# Display the map
m.centerObject(neon, 8)
display(m)
```
